```python
vocab = set()

def tokenizer(s:str) -> list[str]:
    return [t.lower().strip(',.') for t in s.split()]


for file_name in ["pos.txt", "neg.txt","test.txt"]:
    with open(file_name) as fp:
        for line in fp:
            for token in tokenizer(line):
                if token not in vocab:
                    vocab.add(token)

print(f"Vocabulary of size = {len(vocab)}")
```

```python
def conditional_probability(vc:set, file_name:str):

    freq = {}
    prob = {}

    reviews = []
    with open(file_name) as fp:
        for line in fp:
            reviews.append(set(tokenizer(line)))


    for v in list(vc):
        for review in reviews:
            if v in review:
                freq[v] = freq.get(v, 0 ) + 1

    for v in list(vc):
        prob[v] = (freq.get(v,0) +1)/ (len(reviews) + len(vc))

    return prob
```

```python
pos_proba = conditional_probability(vocab, "pos.txt")
neg_proba = conditional_probability(vocab, "neg.txt")
```

```python
sample = """
Malzeme kalitesi son derece düşük, ilk haftanın sonunda çatlamaya başladı.
"""
```

```python
sample_tokens = list(set(tokenizer(sample)))
```

```python
for t in sample_tokens:
    print(f"{t} -> {pos_proba[t]} {neg_proba[t]}")
```

```python
ppos, pneg = 0.5, 0.5

for t in sample_tokens:
    ppos *= pos_proba[t]
    pneg *= neg_proba[t]


print(f"{ppos:.10f}")
print(f"{pneg:.10f}")
```

```python
from math import log, exp

log_ppos, log_pneg = log(0.5), log(0.5)

for t in sample_tokens:
    log_ppos += log(pos_proba[t])
    log_pneg += log(neg_proba[t])


print(f"{log_ppos:.10f}")
print(f"{log_pneg:.10f}")
```
