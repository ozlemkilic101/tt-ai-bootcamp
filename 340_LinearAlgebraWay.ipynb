```python
class Vocabulary:
    def __init__(self):
        self.s2i = {"oov": 0}
        self.i2s = ["oov"]

    def add_word(self, word:str):
        if word in self.s2i:
            return

        self.s2i[word] = len(self.i2s)
        self.i2s.append(word)

    def __len__(self):
        return len(self.i2s)

    def get_word_id(self, word:str)->int:
        return self.s2i.get(word, 0)


vocab = Vocabulary()

def tokenizer(s:str) -> list[str]:
    return [t.lower().strip(',.') for t in s.split()]


for file_name in ["pos.txt", "neg.txt"]:
    with open(file_name) as fp:
        for line in fp:
            for token in tokenizer(line):
                vocab.add_word(token)

print(f"Vocabulary of size = {len(vocab)}")
```

```python
dt = np.zeros((100, len(vocab)))
```

```python
dt.shape
```

```python
doc_id = 0
for file_name in ["pos.txt", "neg.txt"]:
    with open(file_name) as fp:
        for line in fp:
            for token in tokenizer(line):
                dt[doc_id][vocab.get_word_id(token)] = 1
dt
```

```python
c = np.array([0.99999]*50+[0.000001]*50)
```

```python
y = -np.log( 1 / c - 1)
```

```python
from numpy.linalg import lstsq
```

```python
beta, *_ = lstsq(dt, y)
```

```python
1/ (1 + np.exp(-np.dot(beta, dt[0,:])))
```
